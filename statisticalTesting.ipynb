```python
import numpy as np 
import pandas as pd 
from scipy import stats
```

```python
df = pd.read_csv('datasets/grades.csv')
df.head()
```

```python
print('There are {} rows and {} columns.'.format(df.shape[0], df.shape[1]))
```

```python
early_finishers = df[pd.to_datetime(df['assignment1_submission']) < '2016']
early_finishers.head()
```

```python
late_finishers = df[pd.to_datetime(df['assignment1_submission']) >= '2016']
late_finishers.head()
```

```python
late_finishers = df[~df.index.isin(early_finishers.index)]
late_finishers.head()
```

```python
print(early_finishers['assignment1_grade'].mean())
print(late_finishers['assignment1_grade'].mean())
```

```python
# students' t-test allows to form the alternative hypothesis ('they are different') as well as null hypothesis ('they are same')
# and then test that null hypothesis
# for this testing we have to choose a significant level as a threshold for how much of a change we're willing to accept 
# this level is called alpha, e.g. let's use a threshold of 0.5 or 5%
# we're using test_ind() function, it's results include t-statistic and a p-value
# it's this latter value, the probabilty indicates the chance between 0 & 1 of our null hypothesis being true    
```

```python
from scipy.stats import ttest_ind

ttest_ind(early_finishers['assignment1_grade'], late_finishers['assignment1_grade'])
```

```python
# 0.18 is greater than 0.5, thus we can't reject null hypothesis
# our null hypothesis was that the two populations are the same 
```

```python
print(ttest_ind(early_finishers['assignment2_grade'], late_finishers['assignment2_grade']))
print(ttest_ind(early_finishers['assignment3_grade'], late_finishers['assignment3_grade']))
print(ttest_ind(early_finishers['assignment4_grade'], late_finishers['assignment4_grade']))
print(ttest_ind(early_finishers['assignment5_grade'], late_finishers['assignment5_grade']))
print(ttest_ind(early_finishers['assignment6_grade'], late_finishers['assignment6_grade']))
```

```python
df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
df2.head()
```

```python
# let's say our critical value is 0.1 or 10%, we'll report 
# when the p-value isn't less than 10% by comapring df1 & df2
# to give evidence that the columns are different

def test_columns(alpha=0.1): 
    num_diff = 0
    for col in df1.columns:
        teststat, pval = ttest_ind(df1[col], df2[col])
        if pval <= alpha:
            print('Col {} is statistically significantly different at alpha={}, pval={}'. format(col, alpha, pval))
            num_diff = num_diff+1
    
    print('Total number of different was {}, which is {}%'. format(num_diff, float(num_diff)/len(df1.columns)*100))

test_columns()            
```
