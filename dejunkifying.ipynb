```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
languages = ['python', 'javascipt', 'c++', 'sql', 'java']
position = np.arange(len(languages))
popularity = [56, 45, 23, 67, 89]

plt.figure(figsize=(10,8))
plt.bar(position, popularity, align='center')
plt.xticks(position, languages)
plt.ylabel('% popularity')
plt.title('top 5 languages for math & data')
```
